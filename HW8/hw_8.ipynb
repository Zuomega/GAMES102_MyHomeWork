```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import sys
from scipy import spatial
import numpy as np
%matplotlib qt5
mpl.use('Qt5agg')
```

```python
def generate_points(num_points):
    vertices = np.random.rand(num_points,2)
    #print(vertices)
    return vertices
```

```python
def itr_once(points, vo, ax):
    cnt = 0
    print(id(points))
    #print(vo.vertices)
    for apregion in vo.point_region:
        #print(apregion)
        if -1 in vo.regions[apregion]:
            print("~~ ", vo.regions[apregion])
            cnt = cnt + 1
            continue
        print("reg: ", points[cnt])
        nx = (np.sum(np.clip(vo.vertices[vo.regions[apregion]][:,0] , 0, 1))) / len(vo.vertices[vo.regions[apregion]][:,0])
        ny = (np.sum(np.clip(vo.vertices[vo.regions[apregion]][:,1],  0, 1))) / len(vo.vertices[vo.regions[apregion]][:,1])
        
        points[cnt,0] = nx
        points[cnt,1] = ny
        cnt = cnt + 1
    #print('v: ',points)
    nvo = spatial.Voronoi(points)
    spatial.voronoi_plot_2d(nvo, ax=ax)
    vo = nvo
    return points, vo
```

```python
pp= generate_points(5)
print(pp)

fig, ax = plt.subplots()
nvo = spatial.Voronoi(pp)

print("nov ", nvo.regions)
print("points ", nvo.points)
print("pointsreg ", nvo.point_region)
print("ve ", nvo.vertices)
spatial.voronoi_plot_2d(nvo, ax=ax)
fig.show()
```
